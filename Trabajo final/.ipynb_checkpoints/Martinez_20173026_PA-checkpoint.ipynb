#### Pregunta 1: 
1.	Utiliza la base de datos de idh distrital de la tarea 4. Escribe una función que retorne un gráfico de dispersión, y que tome como parámetros el dataframe, el eje x, eje y y el color. Prueba tu función. 

```python
import pandas as pd
import altair as alt
```

```python
distrito = pd.read_csv('idh_distrital_2019.csv')
```

```python
distrito.columns
```

```python
distrito.head(5)
```

```python
def grafico_dispersion(dataframe, ejex, ejey, color):
    distrital = alt.Chart(dataframe, width=800,height=600).mark_circle(opacity = 0.3, color = color).encode(
        x = alt.X(ejex),
        y = alt.Y(ejey,scale=alt.Scale(zero=False)),
        size = alt.Size(ejey),
        tooltip = [alt.Tooltip('nombre', title = 'Distrito'),
               alt.Tooltip(ejex),
               alt.Tooltip(ejey)]).interactive().properties(
        title={
            "text": ["Gráfico de dispersión"], 
            "subtitle": ["A nivel distrital", "2019"],
            "color": "Blue",
            "subtitleColor": "Black"})
    return distrital
```

```python
distrito.columns
```

```python
grafico_dispersion(distrito, 'idh','educacion', 'orange')
```

#### Pregunta 3: 
3.	Realiza un cuarto gráfico con la base de datos del trabajo final, así como una narrativa que explique el gráfico. El tipo de gráfico tiene que ser diferente a los ya presentados en el trabajo final. 

```python
df_positivos = pd.read_csv( "positivos_covid.csv", sep = ";" ,encoding = 'UTF-8')
```

```python
df_positivos.columns
```

```python
df_positivos.shape
```

```python
# Eliminando los valores missings
df_positivos = df_positivos.dropna()
```

```python
df_positivos.isnull().sum(axis = 0)
```

```python
df_positivos.dtypes
```

```python
df_positivos.shape
```

```python
# Cambiando de formato a las variable de fechas
df_positivos["FECHA_RESULTADO"] = df_positivos["FECHA_RESULTADO"].astype(int).astype(str)

df_positivos["FECHA"] =  pd.to_datetime( df_positivos["FECHA_RESULTADO"],
                                         format= '%Y%m%d', errors = "coerce")
```

```python
#Fecha de resultado en mes
df_positivos['MES'] = df_positivos['FECHA'].dt.strftime('%Y%m').astype(int) #YYYYMM
```

```python
df_positivos.head()
```

```python
mdios = (df_positivos[df_positivos["DEPARTAMENTO"] == "MADRE DE DIOS"])
```

```python
mdios.head()
```

```python
mdios.shape
```

```python
mdios2 = mdios[mdios["MES"] >= 202101]
```

```python
mdios2.shape
```

```python
mdios2['infectados'] = 1
```

```python
alt.Chart(mdios2).mark_bar().encode(
    x='sum(infectados)',
    y='EDAD',
    color='METODODX'
).interactive().properties(
        title={
            "text": ["PRUEBAS DE DESCARTE DE COVID-19 SEGÚN EDAD DEL PACIENTE"], 
            "subtitle": ["Madre de Dios", "2021"],
            "color": "Blue",
            "subtitleColor": "Black"})
```

En el gráfico superior se puede observar que existe un uso intensivo de pruebas de antígenos en la población que posee de 30 a 60 años de edad; asimismo, el uso intensivo de pruebas moleculares (PCR) y rápicas (PR) se aplica más en personas de 28 a 43 años de edad. De forma general, se observa que durante el año 2021, en Madre de Dios, las pruebas más utilizadas son la prueba de antígenos y las menos usadas son las pruebas rápidas.
