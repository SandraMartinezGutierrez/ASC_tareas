4. Limpiando las bases de datos

```python
# Eliminando los valores missings

df_positivos.shape
```

```python
# cuántos nulos tienen la base 

df_positivos.isnull().sum(axis = 0)
```

```python
df_positivos = df_positivos.dropna()
```

```python
df_positivos.isnull().sum(axis = 0)
```

```python
#Se eliminaron los valores missings
df_positivos.shape
```

```python
# Cambiando de formato a las variable de fechas
df_positivos["FECHA_RESULTADO"] = df_positivos["FECHA_RESULTADO"].astype(int).astype(str)

df_positivos["FECHA"] =  pd.to_datetime( df_positivos["FECHA_RESULTADO"],
                                         format= '%Y%m%d', errors = "coerce")
```

```python
df_positivos.head()
```

```python
df_positivos["EDAD"] = df_positivos["EDAD"].astype(int)
```

```python
df_positivos.head()
```

```python
#Dummies para metodo de prueba y sexo

df_positivos = pd.get_dummies(df_positivos, columns = ['METODODX',"SEXO"])
```

```python
df_positivos.head()
```

```python
#Fecha de resultado en mes
df_positivos['MES'] = df_positivos['FECHA'].dt.strftime('%Y%m').astype(int) #YYYYMM
```

```python
df_positivos.dtypes
```

```python
df_positivos.head()
```

```python
df_positivos_consolidado = df_positivos.groupby(['MES','DEPARTAMENTO']).agg( 
    {'METODODX_AG':['sum'],
     'METODODX_PCR':['sum'],
     'METODODX_PR':['sum'],
     'SEXO_FEMENINO':['sum'],
     'SEXO_MASCULINO' : ['sum']} )
```

```python
df_positivos_consolidado.head()
```

```python
df_positivos_consolidado.columns = ["antigenos","molecular","prueba_rapida","femenino","masculino"]
df_positivos_consolidado.reset_index(inplace=True)
```

```python
df_positivos_consolidado.head()
```

```python
df_positivos_mensual = df_positivos_consolidado.groupby(["MES"]).sum()
df_positivos_mensual.reset_index(inplace=True)
```

```python
df_positivos_mensual.head()
```

#### Gráfico 1 

```python
plt.rcParams.update({'font.size': 15})  #configurando el tamaño de letra para los gráficos
plt.rcParams["figure.figsize"] = (25,10)
df_positivos_mensual.plot(x = "MES", y = ["antigenos","molecular","prueba_rapida"], kind ="bar",stacked =True)

plt.xticks(rotation = 45) #labels con rotacion
plt.title("Evolución de la Positividad Mensual por tipo de prueba")

#Añadiendo etiquetas con el número de contagios totales por prueba positiva
x = df_positivos_mensual["MES"]
y = df_positivos_mensual["antigenos"] + df_positivos_mensual["molecular"] + df_positivos_mensual["prueba_rapida"]

#Para mover la ubicación de las etiquetas con los datos totales
fx  = 0.99
fy = 0.7
for i, v in enumerate(x):
    plt.text(i*fx, y[i]*fy,str(int(y[i])), color='blue', backgroundcolor= "white",fontsize = 12 ,rotation = 0)
    
plt.show()
```

En este gráfico se puede observar que hasta finales del año 2020, la cantidad de pruebas rápidas que se utilizaban para descartar los casos positivos de COVID-19 eran utilizadas en mayor cantidad que la prueba molecular. Sin embargo, desde el año 2021, se comenzaron a emplear en mayor proporción las pruebas de antígenos, y se estuvo reduciendo paulatinamente el número de pruebas rápidas.

### Gráfico 3

```python
#Sumando los infectados por sexo (masculino y femenino) para tener el total de infectados
df_positivos_dep["infectados_totales"] = df_positivos_dep["femenino"] + df_positivos_dep["masculino"]

#Haciendo un ranking de los departamentos por numero de infectados
df_positivos_dep["rank"] = df_positivos_dep["infectados_totales"].rank(ascending = False)
```

```python
df_positivos_dep.head(5)
```

```python
# Se separan los departamentos en función a los 10 primeros con mas infectados
df_positivos_dep["DEPARTAMENTO_CORREGIDO"] = df_positivos_dep.apply(lambda x: x["DEPARTAMENTO"] if x["rank"] < 10 else "OTROS",axis = 1 )
```

```python
df_positivos_dep.head(5)
```

```python
#Se consolidas nuevamente y se vuelve a ordenar el dataframe
df_positivos_dep2= df_positivos_dep[["DEPARTAMENTO_CORREGIDO","infectados_totales"]].groupby("DEPARTAMENTO_CORREGIDO").sum().reset_index()
```

```python
df_positivos_dep2.head(10)
```

```python
df_positivos_dep2 = df_positivos_dep2.sort_values(by=['infectados_totales'], ascending=False)
```

```python
df_positivos_dep2.head(10)
```

```python
#Seteamos los valores de x,y y explote (explode es para que el grafico resalte)
x = df_positivos_dep2["DEPARTAMENTO_CORREGIDO"]
y = df_positivos_dep2["infectados_totales"]
explode = np.zeros( len(x) )  #resaltar los numeros mayores
explode[0:3] = 0.1 # explode hasta los 3 primeros registros
```

```python
plt.rcParams["figure.figsize"] = (15,7)


fig, ax = plt.subplots()
ax.pie(y, explode=explode, labels=x, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax.set_title("Proporción de Infectados por regiones en el 2021 excluyendo Lima")

plt.show()
```

De acuerdo a este gráfico, las regiones en donde se concentra la mayor cantidad de infectados por COVID-19 son las regiones de Arequipa (8.8%) y Callao (8.7%).
